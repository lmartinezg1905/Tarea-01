```
#Ejercicio 4 

print("Este programa emula una calculadora")

def Suma(u, v):
  suma = u + v
  print(u, "+", v, "=", suma)
def Resta(u, v):
  resta = u - v
  print(u, "-", v, "=", resta)
def Multiplicacion(u, v):
  multi = u * v
  print(u, "*", v, "=", multi)
def Division(u, v):
  while v == 0:
    v = float(input("Introduzca un divisor diferente de cero: "))
  divi = u / v
  print(u, "/", v, "=", divi)
print("Seleccione una de las siguientes funciones: a) Suma, b) Resta, c) Multiplicacion, d) Division ")
x = input("Funcion seleccionada: ")
if x == "a":
  print("SUMA")
  u = float(input("Introduzca el primer numero: "))
  v = float(input("Introduzca el segundo numero: "))
  Suma(u, v)
elif x == "b": 
  print("RESTA")
  u = float(input("Introduzca el primer numero: "))
  v = float(input("Introduzca el segundo numero: "))
  Resta(u, v)
elif x == "c":
  print("MULTIPLICACION")
  u = float(input("Introduzca el primer numero: "))
  v = float(input("Introduzca el segundo numero: "))
  Multiplicacion(u, v)
elif x == "d":
  print("Division")
  u = float(input("Introduzca el primer numero: "))
  v = float(input("Introduzca el segundo numero: "))
  Division(u, v)
else:
  print("No ha seleccionado ninguna operacion")
```
