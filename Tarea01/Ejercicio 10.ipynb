```
#Ejercicio 10

print("Este programa calcula a partir de un mismo radio la formula el area de un circulo y el volumen de un cilindro")

r = float(input("Da el valor de radio: "))
h = float(input("Da el valor de altura: "))
pi = 3.1416

area = pi * r**2

volumen = pi * r**2 * h

print("El resultado del area es: ", area)
print("El resultado del volumen es: ", volumen)

```
